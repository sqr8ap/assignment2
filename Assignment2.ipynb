1. This paper is about data tidying, or restructuring datasets to allow for easy cleaning.
2. The "tidy data standard" is intended to offer a universal starting point for data exploration and analysis and to simplify the development of compatible analysis tools. Datasets come in many different formats, some of which can be very difficult to work with. The tidy structure allows for an easier and more straightforward cleaning process.
3. Sentence 1: Datasets in tidy format all have the same structure and are therefore very similar, but untidy datasets do not share structural characteristics and thus can differ in many more ways. Untidy datasets do not share some structural common denominator in the way that tidy datasets do, and that dissimilarity opens the door to countless potential differences. Sentence 2: When we are working with a dataset, it is generally easy to identify which elements are observations and which elements are variables, especially when we have the context of the data and understand what it is aiming to capture. However, defining a broadly applicable rule or convention for what is a variable and what is an observation is more difficult, as differences in structure present constraints that are not present when looking at one dataset in isolation.
4. The term 'value' describes the actual data points in a dataset. Values are generally either strings or numbers. Each value in a dataset maps to both a 'variable' and an 'observation.' Variables contain all values that measure the same underlying attribute across units, whereas observations contain all values measured on the same unit across attributes.
5. In a tidy dataset, the following is true: 1) each variable forms a column, 2) each observation forms a row, and 3) each type of observational unit forms a table. Tidy datasets can be categorized as third normal form.
6. The five most common problems with messy datasets are as follows: 1) column headers are values rather than variable names, 2) multiple variables are stored in one column, 3) variables are stored in both rows and columns, 4) multiple types of observational units are stored in the same table, and 5) a single observational unit is stored in multiple tables. Table 4 is considered messy because the column headers are values rather than variable names. Melting a dataset involves turning columns into rows and is often referred to as stacking. Melting a dataset eliminates the issue of having column headers that reflect values rather than variable names.
7. Table 11 is messy because there are columns for dates, but dates are values.
```python
import pandas as pd
import numpy as np
import os
```

```python
os.chdir('/content/assignment2/data/')
df = pd.read_csv('airbnb_hw.csv',low_memory=False)
```

```python
# Cleaning the Price variable

var = 'Price'
df[var].str.replace(',','') # here I'm replacing the commas with empty strings
                            # to get rid of them before typecasting
df[var] = pd.to_numeric(df[var], errors='coerce') # here I'm coercing the Price values to numeric

print("After cleaning, there are ", sum(df[var].isnull()), " null values.")
```

```python
# Cleaning the Type variable

df = pd.read_csv('sharks.csv',low_memory=False)

var = 'Type'
# getting an idea of what the variable looks like:
print(df[var].unique())
print(df[var].value_counts())
print(sum(df[var].isnull()))

# replacing Boatomg, Boat, Boating with Watercraft
df[var] = df[var].replace(['Boatomg','Boat','Boating'],'Watercraft')
print(df[var].unique())

# replacing Unconfirmed, Unverified, Under investigation with Questionable
df[var] = df[var].replace(['Unconfirmed','Unverified','Under investigation'],'Questionable')
print(df[var].unique())

# my approach preserved the main categories that values can take on in this
    # column of the dataset; i did not convert anything to nan

# a different approach to cleaning this variable could include coding provoked
    # as 1 and unprovoked as 0 with nan for all other values:
df[var] = df[var].replace('Unprovoked',0)
df[var] = df[var].replace('Provoked',1)
df[var] = df[var].replace(['Questionable','Watercraft','Invalid','Sea Disaster'],np.nan)
print(df[var].value_counts())
print(sum(df[var].isnull()))
```

```python
# Cleaning the WhetherDefendantWasReleasedPretrial variable

url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
df = pd.read_csv(url,low_memory=False)
```
